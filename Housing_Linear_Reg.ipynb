# Load Data

```python
import pandas as pd 
```

```python
cali_dataset = pd.read_csv("California_Houses.csv")
```

```python
cali_dataset
```

- medv will be the y-variable(target)
- everything else is the x-variables(features)

# Data Split

```python
Y = cali_dataset.Median_House_Value
Y
```

- separate variable from features

```python
X = cali_dataset.drop(['Median_House_Value'], axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

- data will be split 80/20
- 80 percent goes to training the data 
- 20 percent goes for the testing data that will be used to make the prediction

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

# Building and using Linear Regression Model

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_test
```

```python
Y_pred = model.predict(X_test)
#creates predictions for house prices based off of our test data
```

```python
Y_test
```

```python
Y_pred
```

```python
print('Intercept:', model.intercept_)
print('MSE: %.2f'% mean_squared_error(Y_test, Y_pred))
print('R2 Score: %.2f'% r2_score(Y_test, Y_pred))
#ytest is actual value
#ypred is predicted value
```
